```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
listing = pd.read_csv('listings.csv')
```

```python
listing
```

```python
sb.countplot(x = 'neighbourhood_group', data = listing)
plt.show()
```

```python
sb.countplot(x = 'neighbourhood', data = listing)
plt.show()
```

```python
sb.barplot(x = 'neighbourhood_group', y ='price', data = listing)
plt.show()
```

```python
sb.barplot(x = 'neighbourhood_group', y ='price', data = listing, ci = False)
plt.show()
```
