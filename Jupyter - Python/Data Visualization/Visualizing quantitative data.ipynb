```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
listing = pd.read_csv('listings.csv')
```

```python
listing
```

```python
plt.hist(listing['price'])
plt.xlabel('price ( in US Dolar)')
plt.show()
```

```python
plt.hist(listing['price'] , bins = np.arange(0,1200, 30))
plt.xlabel('price ( in US Dolar)')
plt.show()
```

```python
plt.scatter(x = listing['price'] , y = listing['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.show()
```

```python
plt.scatter(x = listing['price'] , y = listing['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.xlim(0,1000)
plt.show()
```

```python
plt.scatter(x = listing['price'] , y = listing['number_of_reviews'] , s = 0.2)
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.xlim(0,1000)
plt.show()
```
